```python
from pathlib import Path
from collections import defaultdict

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
GDRIVE = ROOT / "data" / "raw" / "pdfs_gdrive"

# Group by (parent folder + filename) — true duplicates have same path
seen = defaultdict(list)
for f in list(GDRIVE.rglob("*.pdf")) + list(GDRIVE.rglob("*.PDF")):
    seen[str(f).lower()].append(f)

for key, paths in seen.items():
    if len(paths) > 1:
        print(f"DUPLICATE: {paths}")

# If glob really returns same path twice, something fishy. Otherwise:
print(f"\nUnique paths: {len(seen)}")
print(f"Total enumerated: {sum(len(v) for v in seen.values())}")
```

```python
from pathlib import Path
from collections import defaultdict

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
GDRIVE = ROOT / "data" / "raw" / "pdfs_gdrive"

# Find files with same parent + name (case-insensitive)
groups = defaultdict(list)
for f in list(GDRIVE.rglob("*.pdf")) + list(GDRIVE.rglob("*.PDF")):
    key = (str(f.parent), f.name.lower())
    groups[key].append(f)

# Are duplicates same content (size identical)?
import hashlib
print("Checking if duplicates are byte-identical...")
identical = 0
different = 0
for key, paths in groups.items():
    if len(paths) > 1:
        sizes = {p.stat().st_size for p in paths}
        if len(sizes) == 1:
            identical += 1
        else:
            different += 1
            print(f"  DIFFERENT SIZES: {paths[0].name} → {sizes}")

print(f"\nGroups with identical size: {identical}")
print(f"Groups with different size: {different}")
```
